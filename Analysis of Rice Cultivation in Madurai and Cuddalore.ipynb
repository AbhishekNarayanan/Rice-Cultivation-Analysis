```python
file=open('raindata.csv')

madurairain=[]
cuddalorerain=[]
for rows in file.readlines():
    row=rows.split(',')

    if(row[1]=="Madurai" and row[16]>='0'):
      madurairain.append((int(row[2]),float(row[16].strip('\n'))))
    if(row[1].strip()=='Cuddalore' and row[2]<='2010' and row[2]>='1997'):
      cuddalorerain.append((int(row[2]),float(row[16].strip('\n'))))
file1=open('MadurairainDataset.csv','w')
file2=open('CuddalorerainDataset.csv','w')
file1.write('Year'+','+'Total Rainfall in cm'+'\n')
file2.write('Year'+','+'Total Rainfall in cm'+'\n')
madurairain=madurairain[:14]
cuddalorerain=cuddalorerain[:14]
for row in madurairain:
    file1.write(str(row[0])+','+str(row[1])+'\n')
for row in cuddalorerain:
    file2.write(str(row[0])+','+str(row[1])+'\n')
rainm=[i[1] for i in madurairain]
rainc=[i[1] for i in cuddalorerain]
mprodperarea=[float(i[2])/float(i[1]) for i in madurai][:14]
cprodperarea=[float(i[2])/float(i[1]) for i in cuddalore][:14]
print(rainm,rainc)
```

```python
#Continuous Random Variables
#Hist - looks like a smooth curve.
#CDF - curvy

%matplotlib inline
import pandas as pd
from pandas.tools.plotting import scatter_matrix

df = pd.read_csv('MaduraiDataset.csv')

print(df.describe())


#Hist plotting
df.hist(bins = 20, column="Area",normed = True, histtype = 'step', figsize= (10,5),xrot =45)

#CDF plotting
df.hist(bins = 20, column="Area", normed = True, cumulative = True, histtype = 'step', figsize= (10,5),xrot=45)

```
